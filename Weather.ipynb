#### 4. Outlier Detection:

```python
def find_outliers(dataframe):
    outliers_dict = {}

    for column in dataframe.columns:
        # Calculate the first and third quartiles (Q1 and Q3) for each column
        Q1 = np.percentile(dataframe[column], 25)
        Q3 = np.percentile(dataframe[column], 75)

        # Calculate the interquartile range (IQR) for each column
        IQR = Q3 - Q1

        # Define the lower and upper bounds to identify outliers for each column
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find outliers using the lower and upper bounds for each column
        outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)][column].tolist()

        # Store outliers in a dictionary
        outliers_dict[column] = outliers

    return outliers_dict

outliers_result = find_outliers(data)

# Display outliers for each column
for column, outliers in outliers_result.items():
    print(f"Outliers in {column}: {outliers}")
  
```

```python
# Columns with outliers
columns_with_outliers = ['Hum_avg', 'Wind_avg', 'Pres_avg']

# Calculate Z-scores for each column
z_scores = abs(zscore(data[columns_with_outliers]))

# Define a threshold for outliers (e.g., z_score > 3)
outlier_threshold = 3

# Identify rows with outliers
outliers = (z_scores > outlier_threshold).any(axis=1)

# Print the number of outliers
print("Number of outliers:", outliers.sum())
```

```python
# Remove outliers from the dataset
data = data[~outliers]

# Now you can use df_no_outliers for your analysis or modeling

# Example: Print the new dataset shape
print("Shape after removing outliers:", data.shape)
```

## Regression Models

```python
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    print(f'Mean Squared Error: {mse}')
    print(f'R-squared: {r2}')
    print(f'Mean Absolute Error (MAE): {mae}')
    print(f'Root Mean Squared Error (RMSE): {rmse}')

#For overfitting and underfitting 
def evaluate_regression_model(y_train, y_train_pred, y_test, y_test_pred):
    """
    Evaluate the performance of a regression model using Mean Squared Error (MSE) on both training and testing sets.

    Parameters:
    - y_train: true labels for the training set
    - y_train_pred: predicted labels for the training set
    - y_test: true labels for the testing set
    - y_test_pred: predicted labels for the testing set

    Prints:
    - Training MSE
    - Testing MSE
    """

    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)

    print(f"Training MSE: {train_mse}")
    print(f"Testing MSE: {test_mse}")
```

### 2. Exponential Smoothing (ETS)

```python
#pip install statsmodels
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Define the train and test split
train_size = int(len(ts) * 0.8)  # 80% for training, 20% for testing
train, test = ts[:train_size], ts[train_size:]

# Fit Holt-Winters Exponential Smoothing model
model_ets = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12)  # Adjust seasonal_periods as needed
results_ets = model_ets.fit()

# Forecast on the test set
forecast_steps = len(test)
forecast_ets = results_ets.forecast(steps=forecast_steps)
```

```python
# Plot original data, fitted values, and forecast
plt.figure(figsize=(12, 6))
plt.plot(train, label='Training Data')
plt.plot(test, label='Test Data')
plt.title('Exponential Smoothing (ETS) Model Forecast for Average Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
# Evaluate the model
mse = mean_squared_error(test, forecast_ets)
r2 = r2_score(test, forecast_ets)
mae = mean_absolute_error(test, forecast_ets)
rmse = np.sqrt(mse)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
